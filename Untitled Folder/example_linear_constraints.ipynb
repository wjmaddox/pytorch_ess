```python
import torch
import math
import matplotlib.pyplot as plt
```

```python
from pytorch_ess.lin_con_elliptical_slice import LinConEllipticalSliceSampler
```

```python
torch.random.manual_seed(2050)
```

setup some constraints

```python
constraints = torch.randn(3, 2)
bounds = torch.randn(3)
```

```python
%pdb
```

draw an initialization that works for us

```python
f_init = torch.lstsq(bounds.unsqueeze(-1), constraints)[0][:2]
```

```python
sampler = LinConEllipticalSliceSampler(constraints, bounds, 1000, f_init)
```

```python
samples, _ = sampler.run()
```

```python
plt.scatter(*samples)
plt.xlim((-2, 2))
plt.ylim((-2, 2))
```

```python
meshgrid = torch.meshgrid(torch.arange(-2., 2., 0.01), torch.arange(-2., 2, 0.01))
```

```python
xx = torch.stack((meshgrid[0].reshape(-1), meshgrid[1].reshape(-1)))
```

```python
ff = ((constraints.matmul(xx) + bounds.unsqueeze(-1)) >= 0).sum(0)
```

```python
f = plt.scatter(*xx, c=ff)
plt.colorbar(f)
```
